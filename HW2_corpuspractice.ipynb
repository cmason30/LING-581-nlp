```python
from nltk.corpus import nps_chat
```

```python
files = ['10-19-20s_706posts.xml',
'10-19-30s_705posts.xml',
'10-19-40s_686posts.xml',
'10-19-adults_706posts.xml',
'10-24-40s_706posts.xml',
'10-26-teens_706posts.xml',
'11-06-adults_706posts.xml',
'11-08-20s_705posts.xml',
'11-08-40s_706posts.xml',
'11-08-adults_705posts.xml',
'11-08-teens_706posts.xml',
'11-09-20s_706posts.xml',
'11-09-40s_706posts.xml',
'11-09-adults_706posts.xml',
'11-09-teens_706posts.xml']
```

```python
# Compile all files together in corpus

text = []

for i in files:
    
    chatrooms = nps_chat.posts(i)
    text.append(chatrooms)
```

```python
import enchant
```

```python
d = enchant.Dict("en_US")
```

```python
# Sort out words found in english dictionary. 

eng_words = []
new_words = []

for ii in text:
    for jj in ii:
        for kk in jj:
            try:
                if d.check(kk):
                    eng_words.append(kk)

                else:
                    new_words.append(kk)
            except ValueError:
                continue
                
                
```

```python
eng_words
```

```python
new_words
```

```python
import re
```

```python
# Remove strings that do not contain letters.

words2 = []

for i in new_words:
    if bool(re.search('[a-zA-Z]', i)):
        words2.append(i)
        
```

```python
# Remove words with numbers in them.

words3 = []

for i in words2:
    if i.isalpha():
        words3.append(i)
        
```

```python
# get full corpus text. 

full_text = []

for ii in text:
    for jj in ii:
        for kk in jj:
            full_text.append(kk)

len(words3)/len(full_text)
```
