```python
import nltk.corpus
import pandas as pd
```

```python
df = pd.read_html('https://www.lexico.com/explore/weird-and-wonderful-words')
```

```python
strange_words = list(df[0][0])
others = ['cockamamie', 'gobbledegook', 'nincompoops']

strange_words.extend(others)
```

```python
arpabet = nltk.corpus.cmudict.dict()

pronounce = {}
for word in strange_words:
    try:
        pronounce[word] = arpabet[word][0]
    except KeyError:
        continue
```
